```python
df_orders['order_purchase_timestamp'] = pd.to_datetime(df_orders['order_purchase_timestamp'], errors = 'coerce')
df_orders.info()
```

```python
df_orders['order_purchase_timestamp'].isnull().value_counts()
```

#### Filtered dataset to delivered orders to model realized customer demand and avoid noise from cancellations.

```python
df_orders = df_orders[df_orders['order_status'] == 'delivered']
```

```python
df_orders.shape
```

```python
df_orders['order_date'] = df_orders['order_purchase_timestamp'].dt.date
df_orders['order_date'].head()
```

#### Aggregated delivered orders to daily level by counting distinct orders per day, creating the base demand time series.

```python
daily_demand = df_orders.groupby('order_date')['order_id'].agg( demand = 'count')
daily_demand = daily_demand.reset_index()
daily_demand.head()
```

#### Ensured continuity of the time series by adding missing dates and imputing zero demand, a necessary step for reliable time series modeling.Â¶

```python
full_dates = pd.date_range(start = daily_demand['order_date'].min(), end = daily_demand['order_date'].max())
full_dates
```

```python
existing_dates = daily_demand["order_date"].unique()
len(existing_dates)
```

```python
missing_dates = full_dates.difference(existing_dates)
missing_dates
```

```python
daily_demand_complete = daily_demand.set_index('order_date').reindex(full_dates).fillna(0).rename_axis("order_date").reset_index()
daily_demand_complete.head(20)
```

#### Time Series Observations:
The daily demand series shows clear temporal structure with visible trends and repeating patterns, indicating suitability for time series forecasting. Occasional spikes suggest potential seasonal or event-driven effects.

```python
daily_demand_complete.plot(x = 'order_date', y = 'demand', figsize = (12,4))
plt.show()
```

```python
demand_2017 = daily_demand_complete[(daily_demand_complete['order_date'] >= '2017-01-01') & (daily_demand_complete['order_date'] <= '2017-12-31')]
demand_2017.head()
```

```python
demand_2017.plot(x = 'order_date', y = 'demand', figsize = (12, 4))
plt.show()
```

```python
demand_2017_Q2 = demand_2017[demand_2017['order_date'].dt.quarter == 2]
demand_2017_Q2
```

```python
demand_2017_Q2.plot(x = 'order_date', y = 'demand', figsize = (12,4))
plt.show()
```
